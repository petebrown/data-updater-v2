```python
import requests
import datetime

def request_json(url):
    r = requests.get(url)
    return r.json()
```

```python
def get_url(n):
    return f'https://www.sofascore.com/api/v1/team/51/events/last/{n}'
```

```python
all_json = []

n = 0
data = request_json(get_url(n))
while 'error' not in data.keys():
    all_json.append(data)
    print(f'Got data for page {n}')
    n += 1
    data = request_json(get_url(n))
```

```python
all_matches = []

for j in all_json:
    for event in j['events']:
        all_matches.append(event)

len(all_matches)
```

```python
all_stats = []

error_ids = []
access_ids = []

error_count = 0
access_count = 0

for m in all_matches:
    id = m['id']
    url = f'https://www.sofascore.com/api/v1/event/{id}/statistics'
    stats = request_json(url)
    if 'error' not in stats.keys():
        all_stats.append({
            'id': id,
            'data': stats
        })
        access_ids.append(id)
        access_count += 1
    else:
        error_ids.append(id)
        error_count += 1
    print(f'Access to {access_count}/{access_count + error_count} matches')

print(f'Accessed {access_count} matches')
print(f'Error for {error_count} matches')
```

```python
all_match_stats = []

for a in all_stats:
    match_stats = {
        'match_id': a['id'],
        'match_stats': a['data']['statistics'][0]['groups'][0]['statisticsItems']
    }
    all_match_stats.append(match_stats)
```

```python
import pandas as pd

pd.DataFrame(all_match_stats).to_csv('sofascore-json/match_stats.csv', index=False)
```

```python
all_matchday_stats = []

error_ids = []
access_ids = []

error_count = 0
access_count = 0

for m in all_matches:
    id = m['id']
    url = f'https://www.sofascore.com/api/v1/event/{id}/lineups'
    stats = request_json(url)
    if 'error' not in stats.keys():
        all_matchday_stats.append({
            'id': id,
            'data': stats
        })
        access_ids.append(id)
        access_count += 1
    else:
        error_ids.append(id)
        error_count += 1
    print(f'Access to {access_count}/{access_count + error_count} matches')

print(f'Accessed {access_count} matches')
print(f'Error for {error_count} matches')
```

```python
pl_stats = pd.DataFrame(all_matchday_stats)

pl_stats.to_csv('sofascore-json/matchday_stats.csv', index=False)
```
