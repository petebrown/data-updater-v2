```python
from bs4 import BeautifulSoup
import requests

player_list = []

player_count = 0

for i in range(0, 721, 20):
    url = f'http://barryhugmansfootballers.com/club/634/{i}'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    players = soup.find_all('td', class_='tableCell1')
    for player in players:
        player_list.append({
            'player_name': player.text,
            'player_link': player.find('a')['href']
        })
        player_count += 1
        print(f'Added {player_count} player: {player.text}')
```

```python
import pandas as pd
df = pd.DataFrame(player_list)

df.player_name = df.player_name.str.strip()

df
```

```python
def get_player_soup(url):
    r = requests.get(url)
    return BeautifulSoup(r.text, 'html.parser').find('div', id='page_content')

df['player_soup'] = df.player_link.apply(get_player_soup)
```

```python
df.to_csv('players.csv', index=False)
```
